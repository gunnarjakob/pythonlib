```python
import matplotlib.pyplot as plt
import numpy as np
import gvpy as gv
%config InlineBackend.figure_format = 'retina'
```

Importing `gvpy` registers the colormaps saved in `cm.py` as regular matplotlib colormaps.

```python
XX, YY = np.meshgrid(np.linspace(0,1,100), np.linspace(0,1,100))
ZZ = np.sqrt(XX**2 + YY**2)
```

```python
plt.figure()
plt.imshow(ZZ, cmap='turbo')
plt.colorbar(label='turbo');
```

```python
plt.figure()
plt.imshow(ZZ, cmap='turbo_r')
plt.colorbar(label='turbo_r');
```

```python
plt.figure()
plt.imshow(ZZ, cmap='ocean4jbm')
plt.colorbar(label='ocean4jbm');
```

```python
plt.figure()
plt.imshow(ZZ, cmap='ocean4jbm_r')
plt.colorbar(label='ocean4jbm_r');
```
